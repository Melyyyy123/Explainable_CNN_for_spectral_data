# Process of Raw Tuning of the sub_architecture

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("raw_tuning.xlsx")
df_sorted = df.sort_values(by='RMSE_train', ascending=False)
```

```python
df2 = pd.read_excel("conv_5D.xlsx")
```

```python
print(df)
print(df_sorted)
```

```python
plt.figure(figsize=(10, 8))

plt.plot(df_sorted['architecture'], df_sorted['RMSE_train'], 'o-', color='black', linewidth=3, label='RMSE_train')
plt.plot(df_sorted['architecture'], df_sorted['RMSE_test'], 'o-', color='green', linewidth=3, label='RMSE_test')

plt.xlabel('Architecture')
plt.ylabel('RMSE')

plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 8))

plt.plot(df_sorted['architecture'], df_sorted['R2_train'], 'o-', color='blue', linewidth=3, label='R_train')
plt.plot(df_sorted['architecture'], df_sorted['R2_test'], 'o-', color='red', linewidth=3, label='R_test')

plt.xlabel('Architecture')
plt.ylabel('R')

plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 8))

X_axis = np.arange(len(df['architecture'])) 

plt.bar(X_axis - 0.2, df['RMSE_train'], 0.4, color='black', label='RMSE_train')
plt.bar(X_axis + 0.2, df['RMSE_test'], 0.4, color='green', label='RMSE_test')

phase_1 = "4CP F Dr 2D (Tegegn)"  
plt.axvline(2.5, color='red', linestyle='--')
plt.text(-1,188,'Paper Models', color='red')  

phase_2 = "G C F 6D"  
plt.axvline(8.5, color='red', linestyle='--')
plt.text(3.5,188,'Add Dense Layer', color='red')

phase_3 = "G 4C F 4D + same"  
plt.axvline(14.5, color='red', linestyle='--')
plt.text(9.5,188,'Add Convolutions', color='red')

plt.text(15,188,'Add Pooling Layer', color='red')

plt.xlabel('Architecture')
plt.ylabel('RMSE')

plt.xticks(X_axis, df['architecture'], rotation=90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 8))

X_axis = np.arange(len(df['architecture'])) 

plt.bar(X_axis - 0.2, df['R2_train'], 0.4, color='purple', label='R_train')
plt.bar(X_axis + 0.2, df['R2_test'], 0.4, color='orange', label='R_test')


phase_1 = "4CP F Dr 2D (Tegegn)"  
plt.axvline(2.5, color='red', linestyle='--')
plt.text(-1,0.91,'Paper Models', color='red')  

phase_2 = "G C F 6D"  
plt.axvline(8.5, color='red', linestyle='--')
plt.text(3.5,0.91,'Add Dense Layer', color='red')

phase_3 = "G 4C F 4D + same"  
plt.axvline(14.5, color='red', linestyle='--')
plt.text(9.5,0.91,'Add Convolutions', color='red')

plt.text(15,0.91,'Add Pooling Layer', color='red')

plt.xlabel('Architecture')
plt.ylabel('R')

plt.xticks(X_axis, df['architecture'], rotation=90)
plt.legend()
plt.show()
```
