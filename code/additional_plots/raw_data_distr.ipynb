# Raw Data Distribution

This notebook serves for inspection of the raw data distribution which contains all the pine wood samples available in PS20191107_gegl.csv. 

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from sklearn.model_selection import train_test_split
```

```python
data_path = '/home/mel/Masterthesis/Masterthesis_code/raw_data/PS20191107_gegl.csv'
```

```python
raw_data = pd.read_csv(f'{data_path}',sep=';', header=0)
```

```python
print(raw_data)
type(raw_data)
```

```python
subset = raw_data[["year", "Origin"]]
print(subset)
grouped = raw_data.groupby('Origin')
```

```python
plt.figure(figsize=(10, 6))
plt.hist(subset['year'], bins=100, color='skyblue', edgecolor='black')
plt.xlabel('Years')
plt.ylabel('Count')
plt.title('Histogram of Years')
plt.grid(True)
plt.show()
```

```python
unique_countries = subset.groupby('Origin')
num_countries = len(unique_countries)
colors = plt.cm.tab10.colors[:num_countries]
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(data=subset, x='year', bins=100, hue='Origin', multiple="stack", palette=colors, edgecolor='black')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True)
plt.show
```
